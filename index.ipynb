# Rename all the columns to lower case

```python
# make column headings all lowercase

df.columns = [i.title() for i in df.columns]
df.columns
```

# Change the Index to be the Line Names

```python
# make 'Linename' the new index

df.set_index('Linename')
```

# Painstakingly change the index back

```python
# to go back to before, drop the new index column added so identical to before

df.reset_index(drop=True)
```

## Create another column 'Num_Lines' that is a count of how many lines pass through a station. Then sort your dataframe by this column in descending order.

```python
# add new column 'Num_lines' = count of lines thru a station, then sort by 'Num_Lines'

df['Num_Lines'] = df['Linename'].map(lambda i: len(i))
df.sort_values('Num_Lines',ascending=False)
```

## Write a function to clean a column name.

```python
def clean(col_name):
    cleaned = col_name.lower()
    return cleaned
clean('C/A')
```

```python
#This is a list comprehension. It applies your clean function to every item in the list.
#We then reassign that to df.columns
#You shouldn't have to change anything here.
#Your function above should work appropriately here.
df.columns = [clean(i) for i in df.columns] 
```

```python
#Checking the output, we can see the results.
df.columns
```

## Compare subway traffic by day of the week. Display this as a graph.

```python
# format the date columm 
df['date'] = pd.to_datetime(df['date'])

# add new column to show the day of the week
df['day_of_week'] = df['date'].dt.dayofweek

# count how many entries/exits for each day of the week
count = df.groupby('day_of_week').sum()

count.plot(kind = 'bar')
```

```python
df.date.dt.dayofweek
```

## Is there more subway traffic on a weekend or a weekday?    Be specific in comparing magnitudes.

```python
# find average per day for the weekend(true) and weekdays(false)
df['weekend'] = df['day_of_week'].map({0:True, 1:False, 2:False, 3:False, 4:False, 5:False, 6:True})
count = df.groupby('weekend').mean()
count
```
